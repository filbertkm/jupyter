```python
import pandas as pd
```

```python
df = pd.read_csv('data/restaurant_inspections.csv')
```

```python
# cleanup zipcodes
df = df.dropna(axis=0, subset=['ZIPCODE'])
df['ZIPCODE'] = df['ZIPCODE'].astype(int).astype(str).str.zfill(5)

df.head()
```

```python
df['DATE'] = pd.to_datetime(df['INSPECTION DATE'])
df = df.query('DATE >= 20220101')
df = df.sort_values(by=['DATE'], ascending=False)
df.head()
```

```python
pd.set_option('display.max_rows', 100)

dfCuisine = df.groupby(by='CUISINE DESCRIPTION').size().reset_index(name='counts')
dfCuisine.sort_values(by='counts', ascending=False)
```

```python
dfBoro = df.groupby(by='BORO').size().reset_index(name='counts')
dfBoro.sort_values(by='counts', ascending=False)
```
